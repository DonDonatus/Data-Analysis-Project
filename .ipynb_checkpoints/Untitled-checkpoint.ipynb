## BOX PLOT
<hr>

The boxplot above is used to help analyse an item's distribution and key statistical measures. The box represents the interquartile range with the median line inside
while the whiskers show the range of data.

```python
dataset.hist(figsize=(10,10),bins=50)
plt.show()
```

## Histogram
<hr>

The histograms in this dataset provides a visual representation of the frequency or count of different values or ranges within that variable.

```python
x=dataset.density
y=dataset.pH
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('density')
plt.ylabel('pH')
plt.show()
```
